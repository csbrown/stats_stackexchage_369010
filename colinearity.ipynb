```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model
import sklearn.mixture
```

```python
thetas = []
js = []
for j in range(-1,60):
    print(j)
    j = 2**(-j)
    js.append(j)
    coefs = []
    for i in range(1000):
        X = np.linspace(0,1,100)
        X2 = X
        y = 5*X + 4
        XX = np.stack((X,X2)).T + np.random.normal(0,j,(100,2))
        lr = sklearn.linear_model.LinearRegression()
        lr.fit(XX,y)
        coefs.append(lr.coef_)
    coefs = np.array(coefs)
    thetas.append((np.mean(coefs[:,0]), np.std(coefs[:,0])))
thetas = np.array(thetas)   
js = np.array(js)
js = -np.log2(js)
    
plt.scatter(coefs[:,0], coefs[:,1])
plt.xlabel("X1 coefficient")
plt.ylabel("X2 coefficient")
plt.show()
```

```python
plt.scatter(js, thetas[:,1],c='b')
plt.ylabel("std of X1 coef in OLS regression")
plt.xlabel("-log_2 of std of added noise")
plt.show()
```
